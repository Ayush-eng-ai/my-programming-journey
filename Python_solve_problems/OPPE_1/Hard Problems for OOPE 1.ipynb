###Check if Two Strings are Anagrams
 Logic: Sort both strings and compare them.


```python
# Define two strings
string1 = "listen"
string2 = "silent"

# Convert both strings to lowercase and sort the characters
sorted_string1 = sorted(string1.lower())
sorted_string2 = sorted(string2.lower())

# Compare the sorted strings
if sorted_string1 == sorted_string2:
  print(f"'{string1}' and '{string2}' are anagrams.")
else:
  print(f"'{string1}' and '{string2}' are not anagrams.")
```

####Count the Frequency of Elements in a List
 Logic: Use a dictionary to store counts or use collections.Counter().


```python
# Method 1: Using a dictionary

# Define a list
my_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Create an empty dictionary to store counts
frequency_dict = {}

# Iterate through the list and count frequencies
for item in my_list:
  if item in frequency_dict:
    frequency_dict[item] += 1
  else:
    frequency_dict[item] = 1

# Print the frequency dictionary
print("Frequency using dictionary:", frequency_dict)
```

```python
# Method 2: Using collections.Counter()

# Import the Counter class
from collections import Counter

# Define a list
my_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Use Counter to count frequencies
frequency_counter = Counter(my_list)

# Print the frequency counter
print("Frequency using Counter:", frequency_counter)
```

###Find the Intersection of Two Lists
 Logic: Use sets to find common elements.

```python
# Define two lists
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

# Convert lists to sets to easily find the intersection
set1 = set(list1)
set2 = set(list2)

# Find the intersection of the two sets
intersection_set = set1.intersection(set2)

# Convert the intersection set back to a list (optional)
intersection_list = list(intersection_set)

# Print the intersection
print("The intersection of the two lists is:", intersection_list)
```

###Find the First Non-Repeating Character in a String
 Logic: Use a dictionary to store character counts and find the first with a count of 1.


```python
# Define a string
my_string = "programming"

# Create a dictionary to store character counts
char_counts = {}

# Iterate through the string to count character frequencies
for char in my_string:
  char_counts[char] = char_counts.get(char, 0) + 1

# Iterate through the string again to find the first non-repeating character
first_non_repeating = None
for char in my_string:
  if char_counts[char] == 1:
    first_non_repeating = char
    break  # Found the first non-repeating character, so exit the loop

# Print the result
if first_non_repeating:
  print("The first non-repeating character is:", first_non_repeating)
else:
  print("There are no non-repeating characters in the string.")
```

###Check if a Number is a Strong Number
 Logic: A number where the sum of the factorial of its digits is equal to the number itself.


```python
import math

def is_strong_number(number):
  """
  Checks if a number is a strong number.

  A strong number is a number where the sum of the factorial of its digits
  is equal to the number itself.
  """
  original_number = number
  sum_of_factorials = 0
  while number > 0:
    digit = number % 10
    sum_of_factorials += math.factorial(digit)
    number //= 10
  return sum_of_factorials == original_number

# Test cases
num1 = 145
num2 = 123

print(f"{num1} is a strong number: {is_strong_number(num1)}")
print(f"{num2} is a strong number: {is_strong_number(num2)}")
```

###Implement a Simple Calculator Using Functions
 Logic: Define functions for addition, subtraction, multiplication, and division, and call them based on user input.


```python
def add(x, y):
  """Adds two numbers."""
  return x + y

def subtract(x, y):
  """Subtracts two numbers."""
  return x - y

def multiply(x, y):
  """Multiplies two numbers."""
  return x * y

def divide(x, y):
  """Divides two numbers."""
  if y == 0:
    return "Error! Division by zero."
  else:
    return x / y

# Demonstrating the calculator functions with predefined values
num1 = 10
num2 = 5

print(f"{num1} + {num2} = {add(num1, num2)}")
print(f"{num1} - {num2} = {subtract(num1, num2)}")
print(f"{num1} * {num2} = {multiply(num1, num2)}")
print(f"{num1} / {num2} = {divide(num1, num2)}")

num3 = 8
num4 = 0
print(f"{num3} / {num4} = {divide(num3, num4)}")
```

###Binary Search Algorithm
 Logic: Divide the list into two halves and repeatedly check which half contains the target value.


```python
def binary_search(arr, target):
  """
  Implements the Binary Search algorithm.

  Args:
    arr: A sorted list of elements.
    target: The element to search for.

  Returns:
    The index of the target element if found, otherwise -1.
  """
  low = 0
  high = len(arr) - 1

  while low <= high:
    # Find the middle index
    mid = (low + high) // 2
    mid_val = arr[mid]

    # Check if the target is at the middle
    if mid_val == target:
      return mid
    # If target is greater, search in the right half
    elif mid_val < target:
      low = mid + 1
    # If target is smaller, search in the left half
    else:
      high = mid - 1

  # Target not found
  return -1

# Test the binary search function
sorted_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target_element = 23

# Perform the binary search
index = binary_search(sorted_list, target_element)

# Print the result
if index != -1:
  print(f"Element {target_element} is found at index {index}")
else:
  print(f"Element {target_element} is not in the list")

target_element_not_found = 10
index_not_found = binary_search(sorted_list, target_element_not_found)

if index_not_found != -1:
    print(f"Element {target_element_not_found} is found at index {index_not_found}")
else:
    print(f"Element {target_element_not_found} is not in the list")
```
