#Question: 1
Find the Largest Among Three Numbers
 Logic: Use if-elif-else conditions to compare three numbers and determine the largest.


```python
num1 = 10 # First number
num2 = 25 # Second number
num3 = 15 # Third number

# Check if num1 is the largest
if num1 >= num2 and num1 >= num3:
    largest = num1
# Check if num2 is the largest
elif num2 >= num1 and num2 >= num3:
    largest = num2
# If neither num1 nor num2 is the largest, then num3 must be the largest
else:
    largest = num3

# Print the largest number
print(f"The largest number among {num1}, {num2}, and {num3} is {largest}")
```

#Question: 2
Check if a Number is Even or Odd
   Logic: Use the modulo operator % to check divisibility by 2.




```python
num = 10 # The number to check

# Check if the number is divisible by 2 with no remainder
if num % 2 == 0:
    # If the remainder is 0, the number is even
    print(f"{num} is an even number.")
else:
    # If the remainder is not 0, the number is odd
    print(f"{num} is an odd number.")
```

#Question: 3
Calculate the Factorial of a Number
 Logic: Use a loop or recursion to multiply numbers from 1 to n.

```python
num = int(input("Enter a number"))
# The number for which the factorial is to be calculated
factorial = 1 # Initialize factorial to 1

# Check if the number is negative
if num < 0:
   print("Factorial does not exist for negative numbers")
# Check if the number is 0
elif num == 0:
   print("The factorial of 0 is 1")
# Calculate factorial for positive numbers
else:
   # Iterate from 1 up to the number
   for i in range(1, num + 1):
       # Multiply factorial by the current number in the loop
       factorial = factorial * i
   # Print the calculated factorial
   print(f"The factorial of {num} is {factorial}")
```

#Question: 4
Check for Prime Number
 Logic: A prime number has no divisors other than 1 and itself. Loop through numbers up to sqrt(n) to check divisibility.


```python
import math # Import the math module for square root calculation

num = int(input("Enter a number ")) # The number to check for primality

# Prime numbers are greater than 1
if num <= 1:
    print(f"{num} is not a prime number")
else:
    is_prime = True # Assume the number is prime initially
    # Loop from 2 up to the square root of the number
    # We only need to check for divisors up to the square root
    for i in range(2, int(math.sqrt(num)) + 1):
        # If the number is divisible by any number in this range, it's not prime
        if (num % i) == 0:
            is_prime = False # Set is_prime to False
            break # No need to check further, break the loop
    # Check the value of is_prime to determine if the number is prime or not
    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

#Question: 5
Sum of Digits of a Number
 Logic: Use a loop to extract digits using % and // operators.




```python
num = 12345 # The number for which to calculate the sum of digits
sum_of_digits = 0 # Initialize a variable to store the sum

# Loop while the number is greater than 0
while num > 0:
    digit = num % 10 # Get the last digit using the modulo operator
    sum_of_digits += digit # Add the digit to the sum
    num = num // 10 # Remove the last digit using floor division

# Print the sum of the digits
print(f"The sum of the digits is: {sum_of_digits}")
```

#Question: 6
Check if a String is a Palindrome
 Logic: Compare the string with its reverse.

```python
text = "Ayush" # The string to check

# Remove spaces and convert to lowercase for accurate comparison
cleaned_text = text.replace(" ", "").lower()

# Reverse the cleaned string
reversed_text = cleaned_text[::-1]

# Check if the cleaned string is equal to its reverse
if cleaned_text == reversed_text:
    print(f"'{text}' is a palindrome.")
else:
    print(f"'{text}' is not a palindrome.")
```

#Question: 6
Fibonacci Series up to N terms
 Logic: Start with 0, 1 and generate terms by summing the previous two

```python
n_terms = 10 # The number of terms to generate

# First two terms of the Fibonacci series
n1, n2 = 0, 1
count = 0 # Initialize a counter

# Check if the number of terms is valid
if n_terms <= 0:
   print("Please enter a positive integer")
elif n_terms == 1:
   print("Fibonacci sequence up to", n_terms, ":")
   print(n1)
else:
   print("Fibonacci sequence:")
   # Loop to generate the remaining terms
   while count < n_terms:
       print(n1)
       # Calculate the next term by summing the previous two
       nth = n1 + n2
       # Update the previous two terms
       n1 = n2
       n2 = nth
       count += 1 # Increment the counter
```

#Question: 7
Find GCD of Two Numbers
 Logic: Use the Euclidean algorithm or the math.gcd() function.


```python
import math # Import the math module to use the gcd function

num1 = 48 # First number
num2 = 18 # Second number

# Calculate the GCD using math.gcd()
gcd = math.gcd(num1, num2)

# Print the result
print(f"The Greatest Common Divisor of {num1} and {num2} is {gcd}")
```

#Question: 8
Armstrong Number Check
 Logic: Sum of digits raised to the power of the number of digits should be equal to the number itself.




```python
num = 153 # The number to check
num_str = str(num) # Convert the number to a string to easily find the number of digits
num_digits = len(num_str) # Get the number of digits

sum_of_powers = 0 # Initialize a variable to store the sum of powers of digits

# Iterate through each digit of the number (as a string)
for digit_char in num_str:
    digit = int(digit_char) # Convert the digit character back to an integer
    sum_of_powers += digit ** num_digits # Add the digit raised to the power of the number of digits to the sum

# Check if the original number is equal to the sum of the powers of its digits
if num == sum_of_powers:
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```
