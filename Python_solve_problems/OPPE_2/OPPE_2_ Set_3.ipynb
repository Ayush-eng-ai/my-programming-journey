###Extract Border Elements from a List
Write a function extract_border_elements that takes a list of integers as input and returns a new list containing only the first and last elements if the list has more than one element.

   *  If the list has only one element, return it as a single-element list.
   *  If the list is empty, return an empty list.
###NOTE:
This is a function type question, you don't have to take input or print the output, just complete the required function definition.

###Example

extract_border_elements([1, 2, 3, 4]) → [1, 4]
extract_border_elements([5]) → [5]
extract_border_elements([]) → []
extract_border_elements([7, 8, 9, 10, 11]) → [7, 11]
---

**Starting code of Above question:

```python
def extract_border_elements(numbers):

    # Check if list is empty
    if len(numbers) == 0:
        return []

    # Check if list has only one item
    elif len(numbers) == 1:
        return [numbers[0]]

    # If list has more than one item
    else:
        # Return first and last element
        return [numbers[0], numbers[-1]]
```

###Absolute Time Difference Between Two Times
Write a function absolute_time_difference that takes two time strings in the format HH:MM as input and returns the absolute time difference between them in the same format.


###NOTE:
 This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

For example, given the times 14:30 and 06:45, the function should return 07:45, because this is the absolute time difference between the two input times.

###HINT:
 You might want to use formatted string with {val:02} for padding time with leading 0 in case of single digit

Examples

* absolute_time_difference('14:30', '06:45')
'07:45'
* absolute_time_difference('06:45', '14:30')
'07:45'
* absolute_time_difference('02:30', '03:10')
'00:40'
* absolute_time_difference('23:59', '00:00')
'23:59'
---
** below image code starting
image.png

```python
def absolute_time_difference(time1, time2):

    # Split first time into hour and minute
    h1, m1 = time1.split(":")

    # Split second time into hour and minute
    h2, m2 = time2.split(":")

    # Convert first time into total minutes
    total1 = int(h1) * 60 + int(m1)

    # Convert second time into total minutes
    total2 = int(h2) * 60 + int(m2)

    # Find absolute difference in minutes
    diff = abs(total1 - total2)

    # Convert minutes back to hours and minutes
    hours = diff // 60
    minutes = diff % 60

    # Return in HH:MM format
    return f"{hours:02}:{minutes:02}"
```

###Transfer amount
Write a function transfer_amount(accounts:dict, sender:str, recipient:str, amount:int) that takes a dictionary of accounts with account numbers as keys and their respective balances as values, the account number of the sender, the recipient's account number, and the transfer amount if the transfer is valid else no transfer should happen and the dictionary is not modified.

###A transanction is valid if:

* both sender and recipient account numbers are present in accounts dictionary.
* the sender's account has sufficient balance
* the amount is non negative
###Note
that this function does not return anything just modifies the accounts dictionary. In case of invalid transfer dictionary is not updated.

Examples
*  accounts = {'12345': 500, '67890': 1000, '98764': 1500}
*  transfer_amount(accounts, '12345', '67890', 1000)
*  accounts # no change as 12345 has less than 1000
>> {'12345': 500, '67890': 1000, '98764': 1500}
* transfer_amount(accounts, '12345', '67890', 400)
* accounts # 400 transfered from 12345 to 67890
>>{'12345': 100, '67890': 1400, '98764': 1500}
* transfer_amount(accounts, '98764', '67890', 1500)
* accounts # 1500 transfered from 98764 to 67890
>>{'12345': 100, '67890': 2900, '98764': 0}
*  transfer_amount(accounts, '98764', '67890', -100)
* accounts # no change as the amount is negative
>>{'12345': 100, '67890': 2900, '98764': 0}
* transfer_amount(accounts, '29732', '67890', 100)
* accounts # no change as 29732 is not present in accounts
>>{'12345': 100, '67890': 2900, '98764': 0}

---

** starting code of image

```python
def transfer_amount(accounts, sender, recipient, amount):

    # Check sender account exists or not
    if sender not in accounts:
        return

    # Check recipient account exists or not
    if recipient not in accounts:
        return

    # Check amount is not negative
    if amount < 0:
        return

    # Check sender has enough balance
    if accounts[sender] < amount:
        return

    # Deduct money from sender account
    accounts[sender] = accounts[sender] - amount

    # Add money to recipient account
    accounts[recipient] = accounts[recipient] + amount
```

###Words with Consecutive Identical Letters
Write a function words_with_consecutive_letters(words) that takes a list of words and returns a list of words that contain at least one pair of consecutive identical letters (e.g., "hello", "apple"). The function should be case-insensitive.

The order of the words in the result should be same as they are present in the input.

###NOTE:
This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

###Examples:
* words = ["hello", "apple", "ball", "test", "cat"] -> ["hello", "apple", "ball"]
>Explanation: "hello" has 'll', "apple" has 'pp', "ball" has 'll'.
* words = ["sky", "fly", "run", "jump"] -> []
>Explanation: No words have consecutive identical letters.

```python
def words_with_consecutive_letters(words):

    # Store final answer
    result = []

    # Check each word one by one
    for word in words:

        # Convert to lowercase for case-insensitive check
        word = word.lower()

        # Check letters one by one
        for i in range(len(word) - 1):

            # Compare current letter and next letter
            if word[i] == word[i + 1]:

                # Add word in result list
                result.append(word)

                # Stop checking same word again
                break

    # Return final list
    return result
```

###Hand Cricket Match Runs
Given two columns of numbers from 0 to 6, separated by space, played by batsman and bowler in a hand cricket match, find the total number of balls played and the total runs scored by the batsman at the end of the innings.

The match ends when the batsman plays the same number as the bowler or plays the same number twice continously.

The number batsman plays will be added to the runs, except when batsman plays 0, the number bowler plays will be added to the runs.

The number played when the game ends is not counted towards the runs.

####Input Format

Each line with numbers played by batsman and bowler separated by space.
####Output Format

Number of balls played and the total runs separated by space.
Example

###Input:

 * 1   4
* 2 3
* 6 2
* 5 1
* 0 2
* 3 3
####Output:

* 6 16
###Explanation

* 1 4 - 1
* 2 3 - 2
* 6 2 - 6
* 5 1 - 5
* 0 2 - 2 (0 played by batsman so 2 is counted as runs)
* 3 3 - out
* Total: 1+2+2+5+2 = 16
###Input:

* 1 4
* 5 3
* 0 3
* 5 1
* 5 4
###Output:

* 5 14
###Explanation

* 1 4 - 1
* 5 3 - 5
* 0 3 - 3 (0 played by batsman so 2 is counted as runs)
* 5 1 - 5
* 5 4 - out (batsman played 5 twice continously)
* Total: 1+5+3+5 = 14

```python
balls = 0
runs = 0
previous_batsman = -1

while True:
    try:
        # Take input of batsman and bowler
        bat, bowl = map(int, input().split())

        # Count every ball
        balls += 1

        # Check if batsman is out
        if bat == bowl or bat == previous_batsman:
            break

        # If batsman plays 0, bowler runs count
        if bat == 0:
            runs += bowl
        else:
            runs += bat

        # Store current batsman number
        previous_batsman = bat

    except EOFError:
        break

# Final output
print(balls, runs)
```

```python
import math

def det_2x2(m) -> float:
    """Determinant of a 2x2 matrix."""
    a, b, c, d = m[0], *m[1]
    return a * d - b * c


def distance(p1: tuple, p2: tuple) -> float:
    """Euclidean distance between two points."""
    x1, y1 = p1
    x2, y2 = p2
    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)


def zcross(A, B, C):
    """Z component of cross product of vectors AB and BC."""
    ax, ay = A
    bx, by = B
    cx, cy = C

    ab = (bx - ax, by - ay)
    bc = (cx - bx, cy - by)

    return det_2x2([ab, bc])


def perimeter(points: list) -> float:
    """Calculate perimeter of polygon."""
    total = 0

    for i in range(len(points)):
        p1 = points[i]
        p2 = points[(i + 1) % len(points)]   # next point, last joins first
        total += distance(p1, p2)

    return total


def bounding_box(points: list) -> tuple:
    """Return bottom-left and top-right coordinates."""
    x_values = []
    y_values = []

    for point in points:
        x_values.append(point[0])
        y_values.append(point[1])

    bottom_left = (min(x_values), min(y_values))
    top_right = (max(x_values), max(y_values))

    return (bottom_left, top_right)


def area(points: list):
    """Calculate area using Shoelace formula."""
    total = 0

    for i in range(len(points)):
        p1 = points[i]
        p2 = points[(i + 1) % len(points)]
        total += det_2x2([p1, p2])

    return abs(total) / 2


def is_convex(points: list) -> bool:
    """Check whether polygon is convex."""
    n = len(points)

    for i in range(n):
        p1 = points[i]
        p2 = points[(i + 1) % n]
        p3 = points[(i + 2) % n]

        if zcross(p1, p2, p3) < 0:
            return False

    return True
```

###Total Size of Image Files
Given a file containing file sizes and file names separated by a comma, find the total size of image files with extensions .jpg, .jpeg, .png, and .gif (case insensitive).

####Assumptions

The file is in the standard format where each line has a file size followed by a file name.
The file size is in bytes.
The file name is in the format filename.extension.
Examples

####Input File

* 2000,file1.jpg
* 890,file2.txt
* 30500,file3.JPEG
* 12000,file4.png
* 40000,file5.gif
* 490,file6.docx
####Output

* 84500
####Explanation

The total size of image files, which is 2000 + 30500 + 12000 + 40000 = 84500

** start code of image

image.png

```python

# This writes the stdin to the input file
import tempfile
import sys

_, filename = tempfile.mkstemp(prefix="case")

with open(filename, 'w') as f:
    f.write(sys.stdin.read())


img_extensions = {'jpg', 'jpeg', 'png', 'gif'}

# Write your code here

total_size = 0

with open(filename, 'r') as f:
    for line in f:
        line = line.strip()

        if line == "":
            continue

        # Split size and filename
        size, file_name = line.split(',')

        size = int(size)

        # Lowercase for case-insensitive check
        file_name = file_name.lower()

        # Get extension after last dot
        ext = file_name.split('.')[-1]

        # Check image extension
        if ext in img_extensions:
            total_size += size

print(total_size)
```
