Middle element from list

Write a function extract_middle_elements(lst:list) that takes a list of integers and returns a new list containing only the middle element if the list has an odd length, or the two middle elements if the list has an even length.

NOTE: This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

Examples

extract_middle_elements([7, 8, 9, 10, 11, 12, 13]) -> [10]
Explanation: The list has an odd length (7), so the middle element is 10.
extract_middle_elements([3, 6, 9, 12, 15, 18]) -> [9, 12]
Explanation: The list has an even length (6), so the two middle elements are 9 and 12.

---
***Starting Code image:
image.png

```python
def extract_middle_elements(lst: list) -> list:
    n = len(lst)  # Get the total number of elements in the list
    if n % 2 == 1:  # Check if the list has an odd length
        middle_index = n // 2  # Calculate the index of the single middle element
        return [lst[middle_index]]  # Return the middle element in a new list
    else:  # If the list has an even length
        # Calculate the indices of the two middle elements
        middle_index_1 = n // 2 - 1
        middle_index_2 = n // 2
        return [lst[middle_index_1], lst[middle_index_2]]  # Return both middle elements in a new list
```

####Shuffle a Three Word Sentence

Write a function shuffle_sentence(sentence, order) that takes a string sentence containing three words separated by spaces and a tuple order specifying the shuffling order.

The function should return a string with the words of the sentence shuffled according to the specified order.

####NOTE: This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

###Examples

sentence = "apple banana orange", order = (0, 2, 1) -> "apple orange banana" Explanation: 0 - apple , 2 - orange, 1 - banana
sentence = "cat dog mouse", order = (2, 1, 0) -> "mouse dog cat" Explanation: 2 - mouse, 1 - dog, 0 - cat

---

***image in starting code

image.png


```python
def shuffle_sentence(sentence: str, order: tuple) -> str:
    # Split the input sentence into individual words
    words = sentence.split(' ')

    # Create a new list to store the shuffled words
    shuffled_words = [''] * len(words)

    # Iterate through the order tuple and place words accordingly
    for i in range(len(order)):
        # The value in 'order[i]' is the original index of the word
        # The index 'i' is where this word should go in the new sentence
        original_word_index = order[i]
        shuffled_words[i] = words[original_word_index]

    # Join the shuffled words back into a single string with spaces
    return ' '.join(shuffled_words)
```

```python
# Test cases for shuffle_sentence function
print(shuffle_sentence("apple banana orange", (0, 2, 1))) # Expected: apple orange banana
print(shuffle_sentence("cat dog mouse", (2, 1, 0)))       # Expected: mouse dog cat
print(shuffle_sentence("one two three", (1, 0, 2)))      # Expected: two one three
print(shuffle_sentence("first second third", (0, 1, 2))) # Expected: first second third (no shuffle)
```

Check if both numbers have the same sign
Write a function same_sign that checks whether two given numbers have the same sign. Consider three cases:

Both numbers are strictly positive.
Both numbers are strictly negative.
Both numbers are zero.
The function should return True if the numbers have the same sign; otherwise, it should return False.

NOTE: This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

Examples

same_sign(10, 20) -> True
Explanation: Both numbers are strictly positive.
same_sign(-5, -15) -> True
Explanation: Both numbers are strictly negative.
same_sign(0, 0) -> True
Explanation: Both numbers are zero.
same_sign(10, -20) -> False
Explanation: The numbers have different signs.
same_sign(0, 10) -> False
Explanation: One number is zero, and the other is strictly positive.

---
***Starting code image

image.png

```python
def same_sign(num1: int, num2: int) -> bool:
    # Check if both numbers are positive, negative, or zero
    if (num1 > 0 and num2 > 0): # Both strictly positive
        return True
    elif (num1 < 0 and num2 < 0): # Both strictly negative
        return True
    elif (num1 == 0 and num2 == 0): # Both zero
        return True
    else: # Different signs or one is zero and the other is not
        return False
```

```python
# Test cases for same_sign function
print(same_sign(10, 20))    # Expected: True
print(same_sign(-5, -15))   # Expected: True
print(same_sign(0, 0))      # Expected: True
print(same_sign(10, -20))   # Expected: False
print(same_sign(0, 10))     # Expected: False
print(same_sign(-10, 0))    # Expected: False
```

Check Palindrome - Advanced
Write a function is_palindrome that takes a string as input and returns True if the string reads the same forward and backward considering only the alphabets and numbers (ignoring spaces, capitalization, and punctuation). Otherwise, it returns False.

Ignore spaces, capitalization, and punctuation while checking.
The function should not print anything.
Complete the function definition only.
Hint: use str.isalnum function to check if a character is alphabet or number

NOTE: This is a function type question, you don't have to take input or print the output, just complete the required function definition.

Example

is_palindrome("A man, a plan, a canal, Panama!") -> True
is_palindrome("racecar") -> True
is_palindrome("hello") -> False
is_palindrome("No lemon, no melon") -> False

---
***starting code above this question

image.png

```python
def is_palindrome(s: str) -> bool:
    # Convert the string to lowercase and filter out non-alphanumeric characters
    filtered_chars = [char.lower() for char in s if char.isalnum()]

    # Join the filtered characters to form a clean string
    cleaned_string = "".join(filtered_chars)

    # Check if the cleaned string is a palindrome by comparing it to its reverse
    return cleaned_string == cleaned_string[::-1]
```

Abbreviate Initials And Sort
Write a Python function that takes a list of full names, processes them to create abbreviated forms in the format "Last, F.M." (where F and M are the initials of the first and middle names), and returns the list sorted alphabetically.

Assume that the names are given in the correct case.

Hint: Use sorted function or list.sort to sort the list

NOTE: This is an I/O type question, you need to write the whole code for taking input and printing the output.

Input Format

First line contains the number of names, n.
Next n lines contain one full name per line.
Output Format

Output the processed names in sorted order, one per line.
Example

Input:

3
John Doe
Alice Johnson
Bob Alan Rickman
Output:

Doe, J.
Johnson, A.
Rickman, B.A.

```python
def abbreviate_and_sort(names):
    # Initialize an empty list to store the processed names
    result = []

    # Loop through each full name in the input list
    for full_name in names:
        # Split the full name into individual words
        name_parts = full_name.split()

        # The last part is always the last name
        last_name = name_parts[-1]

        # Initialize a string to store initials (e.g., 'F.M.')
        initials = ""

        # Iterate through all parts except the last name to get initials
        for i in range(len(name_parts) - 1):
            # Take the first letter of each part and append with a dot
            initials += name_parts[i][0] + "."

        # Format the name as "Last, F.M."
        # If there are no initials, it will just be "Last, " which needs to be handled
        # The current logic will produce "Last, F." for a two-part name (John Doe -> Doe, J.)
        # and "Last, F.M." for a three-part name (Bob Alan Rickman -> Rickman, B.A.)
        # which matches the examples.
        result.append(f"{last_name}, {initials.strip()}") # .strip() removes any leading/trailing whitespace if initials was empty

    # Sort the list of abbreviated names alphabetically
    result.sort()

    return result
```

```python
# Main part for input and output
if __name__ == '__main__':
    # Read the number of names
    n = int(input())

    # Read n names into a list
    input_names = []
    for _ in range(n):
        input_names.append(input())

    # Process and sort the names using the defined function
    sorted_abbreviated_names = abbreviate_and_sort(input_names)

    # Print the sorted abbreviated names, one per line
    for name in sorted_abbreviated_names:
        print(name)
```

```python
#### Second type of code only change the  parameter



def employee_task_analysis(employee_list, task):

    # ---------------------------------------------------
    # 1. Find total tasks completed by all employees
    # ---------------------------------------------------
    if task == "total_tasks_completed":

        total = 0   # start total from 0

        # check every employee one by one
        for emp in employee_list:

            # add that employee's completed tasks
            total = total + emp['tasks_completed']

        # return final total
        return total


    # ---------------------------------------------------
    # 2. Find most efficient employee
    # efficiency = tasks_completed / hours_worked
    # If tie -> choose employee with more tasks
    # ---------------------------------------------------
    elif task == "most_efficient_employee":

        best_id = None      # store best employee id
        best_ratio = -1     # store highest efficiency
        best_tasks = -1     # store tasks for tie case

        # check every employee
        for emp in employee_list:

            # calculate efficiency
            ratio = emp['tasks_completed'] / emp['hours_worked']

            # if current employee is better
            if ratio > best_ratio:

                best_ratio = ratio
                best_tasks = emp['tasks_completed']
                best_id = emp['employee_id']

            # if efficiency same, compare tasks
            elif ratio == best_ratio:

                if emp['tasks_completed'] > best_tasks:

                    best_tasks = emp['tasks_completed']
                    best_id = emp['employee_id']

        # return best employee id
        return best_id


    # ---------------------------------------------------
    # 3. Sort by hours worked (high to low)
    # If same hours -> more tasks first
    # Return only employee ids
    # ---------------------------------------------------
    elif task == "sort_by_hours_worked":

        # sorted() sorts list
        # minus sign means descending order
        sorted_list = sorted(
            employee_list,
            key=lambda emp: (-emp['hours_worked'],
                             -emp['tasks_completed'])
        )

        result = []   # store ids only

        # take ids from sorted employees
        for emp in sorted_list:
            result.append(emp['employee_id'])

        return result


    # ---------------------------------------------------
    # 4. Employees who completed 20 or more tasks
    # Keep original order
    # ---------------------------------------------------
    elif task == "productive_team_members":

        result = []

        # check every employee
        for emp in employee_list:

            # if tasks are 20 or more
            if emp['tasks_completed'] >= 20:

                # add employee id
                result.append(emp['employee_id'])

        return result
```

###Simple Stemmer

Given a multiline text file containing lowercase words separated by spaces or newlines, stem each word (remove certain suffixes) according to the list of suffixes given in the suffixes list and print the contents in the same format.

Do not change the line or order of the words.

suffixes = [

    'wards', 'ations', 'ation', 'tions', 'tion', 'asions',
    'asion', 'sions', 'sion', 'ment', 'ness', 'ship',
    'hood', 'able', 'ible', 'less', 'ward', 'wise', 'ion', 'ity', 'age',
    'ize', 'ise', 'ify', 'ate', 'ful', 'ous', 'ish', 'ive', 'ing', 'ers', 'er',
    'or', 'ty', 'en', 'ic', 'al', 'ly'
]

Remove the first suffix from the list of suffixes that matches a given word. For example for the word "education" the suffix "ation" and "tion" matches but "ation" is present first in the suffixes list thus the word is stemmed as "educ".

###NOTE:
 This is a file-in-stdout type question where the input is read from the file and the output is printed in the standard output.

```python
import tempfile
import sys

_, filename = tempfile.mkstemp(prefix="case")

with open(filename, 'w') as f:
    f.write(sys.stdin.read())


suffixes = [
    'wards', 'ations', 'ation', 'tions', 'tion', 'asions',
    'asion', 'sions', 'sion', 'ment', 'ness', 'ship',
    'hood', 'able', 'ible', 'less', 'ward', 'wise',
    'ion', 'ity', 'age', 'ize', 'ise', 'ify', 'ate',
    'ful', 'ous', 'ish', 'ive', 'ing', 'ers', 'er',
    'or', 'ty', 'en', 'ic', 'al', 'ly'
]


def stem_word(word):
    for suffix in suffixes:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word


def main():

    result = []

    # read file input
    with open(filename, 'r') as f:
        lines = f.readlines()

    for line in lines:

        words = line.strip().split()

        new_words = []

        for word in words:
            new_words.append(stem_word(word))

        result.append(' '.join(new_words))

    print('\n'.join(result))


if __name__ == "__main__":
    main()
```
