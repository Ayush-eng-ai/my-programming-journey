# Python Built in Modules  with uses:
---
---

| Module   | Main Use           | What You Can Do                                               |
| -------- | ------------------ | ------------------------------------------------------------------ |
| math     | Calculations       | square root, power, factorial, trigonometry, rounding, pi use      |
| random   | Random values      | random number, password generate, shuffle list, random choice      |
| string   | Text constants     | alphabets, digits, punctuation, text checking                      |
| os       | File/folder system | folder create, file delete, rename, current path, automation       |
| sys      | Python system info | Python version, command line input, exit program, memory/path info |
| time     | Delay/time         | delay (`sleep`), timestamp, timer create                           |
| datetime | Date & time        | current date, age calculator, date difference, formatting          |
| re       | Pattern matching   | email validate, mobile check, search text pattern, replace text    |
| json     | API / JSON data    | dictionary to JSON, JSON to dict, API data handle                  |
| csv      | Excel/CSV files    | CSV read, CSV write, report create, dataset handle                 |


```python
# Uses
# =========================
# 1. math module examples
# =========================
import math

# square root
print(math.sqrt(49))          # 7.0

# power
print(math.pow(2, 5))         # 32.0

# factorial
print(math.factorial(5))      # 120

# trigonometry
print(math.sin(math.radians(90)))   # 1.0

# rounding
print(math.ceil(4.2))         # 5
print(math.floor(4.9))        # 4

# pi use
radius = 7
area = math.pi * radius ** 2
print(area)


# =========================
# 2. random module examples
# =========================
import random

# random number
print(random.randint(1, 100))

# password generate
chars = "abc123XYZ"
password = "".join(random.choice(chars) for _ in range(6))
print(password)

# shuffle list
nums = [1, 2, 3, 4, 5]
random.shuffle(nums)
print(nums)

# random choice
colors = ["red", "blue", "green"]
print(random.choice(colors))


# =========================
# 3. string module examples
# =========================
import string

# alphabets
print(string.ascii_letters)

# digits
print(string.digits)

# punctuation
print(string.punctuation)

# text checking
text = "Hello123"
print(text.isalnum())     # True


# =========================
# 4. os module examples
# =========================
import os

# current path
print(os.getcwd())

# folder create
os.mkdir("demo_folder")

# rename folder
os.rename("demo_folder", "new_folder")

# file delete (if file exists)
# os.remove("test.txt")

# automation (list files)
print(os.listdir())


# =========================
# 5. sys module examples
# =========================
import sys

# Python version
print(sys.version)

# command line input
print(sys.argv)

# memory/path info
print(sys.path)

# exit program
# sys.exit()


# =========================
# 6. time module examples
# =========================
import time

# timestamp
print(time.time())

# delay
print("Wait...")
time.sleep(2)
print("Done")

# timer create
start = time.time()
for i in range(100000):
    pass
end = time.time()
print("Execution Time:", end - start)


# =========================
# 7. datetime module examples
# =========================
from datetime import datetime, date

# current date
print(date.today())

# formatting
now = datetime.now()
print(now.strftime("%d-%m-%Y %H:%M:%S"))

# age calculator
birth = date(2005, 8, 15)
today = date.today()
print(today.year - birth.year)

# date difference
d1 = date(2026, 1, 1)
d2 = date(2026, 4, 24)
print((d2 - d1).days)


# =========================
# 8. re module examples
# =========================
import re

text = "Contact: ayush@gmail.com, phone: 9876543210"

# email validate
print(re.findall(r'\S+@\S+', text))

# mobile check
print(re.findall(r'\d{10}', text))

# search text pattern
print(re.search("Contact", text))

# replace text
print(re.sub("Contact", "Info", text))


# =========================
# 9. json module examples
# =========================
import json

data = {"name": "Ayush", "age": 20}

# dictionary to JSON
json_data = json.dumps(data)
print(json_data)

# JSON to dict
python_data = json.loads(json_data)
print(python_data)

# API data handle style
print(python_data["name"])


# =========================
# 10. csv module examples
# =========================
import csv

# CSV write
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Marks"])
    writer.writerow(["Ayush", 95])

# CSV read
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# report create / dataset handle
with open("students.csv", "a", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Rohit", 88])
```
